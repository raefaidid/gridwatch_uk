```python
import pandas as pd

#ingest raw data
df = pd.read_csv('data/gridwatch_110514_240914.csv')
```

```python
dim_energy_sources = pd.melt(df_sample, id_vars=['id'], var_name='source_name',value_vars=['coal', 'nuclear', 'ccgt',
                            'wind', 'pumped', 'hydro', 'biomass', 'oil', 'solar', 'ocgt'])\
                            .sort_values('id')\
                            .reset_index(drop=True)\
                            .rename(columns={'id' : 'energy_source_id',
                                             'source_name': 'source_name',
                                             'value' : 'output'})
                            
dim_energy_sources
```

```python
import numpy as np

def check_direction(df):
   if df.flow > 0 :
      return 'import'
   elif df.flow == 0:
      return np.nan
   else:
      return 'export'

dim_interconnectors = pd.melt(df_sample, id_vars='id',
                              var_name='shortname',value_vars=['french_ict', 'dutch_ict', 'irish_ict',
                                                                         'ew_ict','nemo', 'ifa2', 'intelec_ict', 'nsl', 'vkl_ict'])\
                                                                            .sort_values('id')\
                                                                            .reset_index(drop=True)\
                                                                            .rename(columns={'value':'flow',
                                                                                             'id' : 'interconnector_id'})
                                                                            
fullname_mapping = {
   'french_ict' : 'French',
   'dutch_ict' : 'Dutch',
   'irish_ict' : 'Irish-Scotland',
   'ew_ict' : 'Irish-Wales',
   'nemo' : 'NEMO',
   'ifa2' : 'French',
   'intelec_ict' : 'French-Channel Tunnel',
   'nsl' : 'Norway',
   'vkl_ict' : 'Norway Viking'
}
                                                                            
dim_interconnectors = dim_interconnectors.assign(
   direction = lambda x: x.apply(check_direction, axis=1),
   fullname = lambda x: x.shortname.map(fullname_mapping)
)

dim_interconnectors
```

```python
dim_flow = df_sample[['id','north_south', 'scotland_england']]\
    .drop_duplicates()\
    .rename(columns={'north_south' : 'north_south_flow',
                     'scotland_england' : 'scotland_england_flow',
                     'id' : 'flow_id'})

dim_flow
```

```python
dim_frequency = df_sample[['id', 'frequency']].rename(columns={'id': 'frequency_id'})

dim_frequency
```

```python
dim_demand = df_sample[['id', 'demand']].rename(columns={'id': 'demand_id'})

dim_demand
```

```python
df.columns = [column.lower().strip() for column in df.columns]

if 'timestamp' in df.columns: df['timestamp'] = pd.to_datetime(df.timestamp)

dim_energy_sources = pd.melt(df, id_vars=['id'], var_name='source_name',value_vars=['coal', 'nuclear', 'ccgt',
                            'wind', 'pumped', 'hydro', 'biomass', 'oil', 'solar', 'ocgt'])\
                            .sort_values('id')\
                            .reset_index(drop=True)\
                            .rename(columns={'source_name': 'source_name',
                                             'value' : 'output'})

dim_energy_sources = dim_energy_sources.assign(energy_source_id = lambda x: x.index)

                       
def check_direction(df):
   if df.flow > 0 :
      return 'import'
   elif df.flow == 0:
      return np.nan
   else:
      return 'export'

dim_interconnectors = pd.melt(df, id_vars='id',
                              var_name='shortname',value_vars=['french_ict', 'dutch_ict', 'irish_ict',
                                                                         'ew_ict','nemo', 'ifa2', 'intelec_ict', 'nsl', 'vkl_ict'])\
                                                                            .sort_values('id')\
                                                                            .reset_index(drop=True)\
                                                                            .rename(columns={'value':'flow'})
                                                                            
fullname_mapping = {
   'french_ict' : 'French ICT',
   'dutch_ict' : 'Dutch ICT',
   'irish_ict' : 'Irish ICT',
   'ew_ict' : 'E-W ICT',
   'nemo' : 'NEMO ICT',
   'ifa2' : 'French ICT 2',
   'intelec_ict' : 'French INTELEC',
   'nsl' : 'Norway ICT',
   'vkl_ict' : 'Viking ICT'
}
                                                                            
dim_interconnectors = dim_interconnectors.assign(
   direction = lambda x: x.apply(check_direction, axis=1),
   fullname = lambda x: x.shortname.map(fullname_mapping),
   interconnector_id = lambda x: x.index
)


dim_flow = df[['id','north_south', 'scotland_england']]\
    .drop_duplicates()\
    .rename(columns={'north_south' : 'north_south_flow',
                     'scotland_england' : 'scotland_england_flow'})

dim_flow = dim_flow.assign(flow_id = lambda x: x.index)

dim_frequency = df[['id', 'frequency']].rename(columns={'id': 'frequency_id'}).assign(frequency_id = lambda x: x.index)

dim_demand = df[['id', 'demand']].rename(columns={'id': 'demand_id'}).assign(demand_id = lambda x: x.index)

dim_time = df[['id','timestamp']].assign(
    time_id = lambda x: x.index,
    year = lambda x: x.timestamp.dt.year,
    month = lambda x: x.timestamp.dt.month,
    day = lambda x: x.timestamp.dt.day,
    hour = lambda x: x.timestamp.dt.hour,
    day_of_week = lambda x: x.timestamp.dt.day_of_week,
    week = lambda x: x.timestamp.dt.isocalendar().week,
)
```

# Read parquet file using duckdb

```python
import duckdb as ddb
```

```python
conn = ddb.connect('raw_gridwatch.db')

conn.execute("""CREATE OR REPLACE TABLE gridwatch AS
             SELECT * FROM read_parquet('data/gridwatch_110514_240914_clean.gzip');""")
```

```python
conn.table('gridwatch').show(max_rows=4)
```

```python
query = """
SELECT
    DATE_PART('year',timestamp) as year,
    DATE_PART('doy',timestamp) as day,
    AVG(coal)
FROM gridwatch
GROUP BY 1, 2
ORDER BY 1,2 ASC;
"""


conn.execute(query).fetch_df()
```

```python
#daily average demand
query = """
SELECT
    strptime(strftime(timestamp, '%Y-%m'),'%Y-%m') as dates,
    ROUND(AVG(demand),2) as average_demand
FROM gridwatch
GROUP BY 1
ORDER BY 1;
"""


conn.execute(query).fetch_df()
```

```python
#daily average frequency
query = """
SELECT
    strptime(strftime(timestamp, '%Y-%m'),'%Y-%m') as dates,
    ROUND(AVG(frequency),2) as average_frequency
FROM gridwatch
GROUP BY 1
ORDER BY 1;
"""


conn.execute(query).fetch_df()
```

```python
#daily peak demand
query = """
SELECT
    strptime(strftime(timestamp, '%Y-%m'),'%Y-%m') as dates,
    ROUND(MAX(frequency),2) as peak_frequency
FROM gridwatch
GROUP BY 1
ORDER BY 1;
"""


conn.execute(query).fetch_df()
```

# Data Modelling V2

```python
df = pd.read_csv('data/gridwatch_110514_240914.csv')

def basic_data_cleaning(df):
    df.columns = [column.lower().strip() for column in df.columns]
    if 'timestamp' in df.columns: df['timestamp'] = pd.to_datetime(df.timestamp)
    return df

df = basic_data_cleaning(df)
```

```python
dim_datetime = df[['timestamp']].drop_duplicates()

dim_datetime = dim_datetime.assign(
    datetime_id = lambda x: x.index,
    year = lambda x: x.timestamp.dt.year,
    month = lambda x: x.timestamp.dt.month,
    day = lambda x: x.timestamp.dt.day,
    hour = lambda x: x.timestamp.dt.hour,
    day_of_week = lambda x: x.timestamp.dt.day_of_week,
    week = lambda x: x.timestamp.dt.isocalendar().week,
)

dim_datetime = dim_datetime[['datetime_id','timestamp', 'year', 'month', 'day', 'hour', 'day_of_week', 'week']]

dim_datetime
```

```python
dim_energy_output_and_flow = df[['id','demand', 'frequency', 'coal', 'nuclear', 'ccgt', 'wind', 'pumped', 'hydro', 'biomass', 'oil', 'solar', 'ocgt',
       'french_ict', 'dutch_ict', 'irish_ict', 'ew_ict', 'nemo', 'other',
       'north_south', 'scotland_england', 'ifa2', 'intelec_ict', 'nsl',
       'vkl_ict']]\
        .drop_duplicates()
        
dim_energy_output_and_flow = dim_energy_output_and_flow.rename(columns={
       'id': 'energy_id',
       'ifa2' : 'french_ict_2',
       'intelec_ict' : 'french_ict_intelec',
       'nsl' : 'norway_ict', 
})

dim_energy_output_and_flow
```

```python
fct_gridwatch = df.reset_index().rename(columns={'index':'fact_id'}).loc[:,['fact_id','id','timestamp']]

fct_gridwatch = fct_gridwatch\
    .merge(dim_energy_output_and_flow[['energy_id']],left_on='id', right_on='energy_id')\
    .merge(dim_datetime[['datetime_id','timestamp']],on='timestamp')\
    .drop(columns=['id','timestamp'])

fct_gridwatch = fct_gridwatch[['fact_id', 'datetime_id', 'energy_id']]

fct_gridwatch
```

```python
dim_datetime.to_parquet('data/dim_datetime.gzip',compression='gzip')
dim_energy_output_and_flow.to_parquet('data/dim_energy_output_and_flow.gzip',compression='gzip')
fct_gridwatch.to_parquet('data/fct_gridwatch.gzip',compression='gzip')
```

## Load data model in Duckdb

```python
import duckdb as ddb
```

```python
conn = ddb.connect('gridwatch.db')

create_schema_query ="""
CREATE SCHEMA warehouse;
"""

conn.execute(create_schema_query)
```

```python
create_fct_table_query ="""
CREATE OR REPLACE TABLE warehouse.fct_gridwatch AS
SELECT * FROM read_parquet('data/fct_gridwatch.gzip');
"""

create_dim_datetime_query ="""
CREATE OR REPLACE TABLE warehouse.dim_datetime AS
SELECT * FROM read_parquet('data/dim_datetime.gzip');
"""

create_dim_energy_output_and_flow_query ="""
CREATE OR REPLACE TABLE warehouse.dim_energy_output_and_flow AS
SELECT * FROM read_parquet('data/dim_energy_output_and_flow.gzip');
"""

conn.execute(create_fct_table_query)

conn.execute(create_dim_datetime_query)

conn.execute(create_dim_energy_output_and_flow_query)
```
